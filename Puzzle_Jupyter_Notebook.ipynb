```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
puzzle_df = pd.read_csv('puzzle.csv', header = None)
```

```python
puzzle_df = puzzle_df.rename(index = str, columns = {0:'col_a', 1:'col_b'})
```

### Insights Found in the Data
The data points seem to refer to coordinates of cities/regions on a worldmap. My hypothesis is that those cities correspond to all (user) cities/region from where we have seen at least one search from the Hopper application, since its creation.

```python
plt.clf()
plt.figure(figsize = (12,6))
plt.plot(puzzle_df.set_index('col_b').sort_index()['col_a'], 'ro')
plt.title('Scatter Plot of Column A against Column B')
plt.xlabel('Column B')
plt.ylabel('Column A')
plt.show()
```

### Exploration of the Data

```python
#Frequency of pair of values
puzzle_agg = puzzle_df.groupby(['col_a', 'col_b']).agg({'col_a':'count'}).rename(str, columns={'col_a':'num_rows'})
```

```python
puzzle_agg.sort_values('num_rows', ascending = False).head(20)
```

```python
#Correlation between the two variables
np.corrcoef(puzzle_df['col_a'], puzzle_df['col_b'])
```

```python
#Distribution of Column A
puzzle_df['col_a'].hist()
plt.show()
```

```python
#Distribution of Column B
puzzle_df['col_b'].hist()
plt.show()
```
